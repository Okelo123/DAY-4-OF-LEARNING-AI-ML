```python
import random

# Simulate coin toss
outcomes = ["H", "T"]
tosses = [random.choice(outcomes) for _ in range(10)]
print("10 coin tosses:", tosses)

# Probability estimation (frequency)
prob_H = tosses.count("H") / len(tosses)
print("Estimated Probability of Heads:", prob_H)
```

```python
import numpy as np
from scipy import stats

data = np.random.randint(50, 100, 20)  # random dataset

print("Data:", data)
print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Standard Deviation:", np.std(data))

# Normal distribution example
mu, sigma = 0, 1
samples = np.random.normal(mu, sigma, 1000)

print("Sample Mean:", np.mean(samples))
print("Sample Std:", np.std(samples))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot histogram of normal distribution
sns.histplot(samples, kde=True, bins=30)
plt.title("Normal Distribution")
plt.show()
```
